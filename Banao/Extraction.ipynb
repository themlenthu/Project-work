## Imports

```python
from selenium import webdriver 
from selenium.webdriver import Chrome
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import TimeoutException
import time
import pandas as pd
```

## Output Dataframe

```python
columns = ['Product Name', 'Price', 'Rating', 'Seller Name']
out_df = pd.DataFrame(columns = columns)
```

## Extraction

```python
url = 'https://www.amazon.in/s?rh=n%3A6612025031&fs=true&ref=lp_6612025031_sar'

options = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=options)
# Set an implicit wait of 20 seconds to allow time for elements to appear before throwing an exception
driver.implicitly_wait(20)
driver.get(url)
```

```python
## Product Name

prod_names = driver.find_elements(By.XPATH,'//span[@class="a-size-base-plus a-color-base a-text-normal"]')
prod_names_text = [i.text for i in prod_names]
out_df['Product Name'] = prod_names_text

## Price

price = driver.find_elements(By.XPATH, '//span[@class="a-price-whole"]')
price_text = []
for i in price:
    if ',' in i.text:
        price_text.append(''.join(i.text.split(',')))
    else:
        price_text.append(i.text)
out_df['Price'] = price_text

## Rating

rating_text = []
rating = driver.find_elements(By.XPATH,'//div[@class="a-row a-size-small"]//span[1]')
for r in rating:
    if r.get_attribute('aria-label'):
        rating_text.append(r.get_attribute('aria-label').split(' ')[0])
out_df['Rating'] = rating_text

## Seller Name

for i in range(out_df.shape[0]):
    out_df.loc[i, 'Seller Name'] = out_df.loc[i, 'Product Name'].split(' ')[0].upper()
```
