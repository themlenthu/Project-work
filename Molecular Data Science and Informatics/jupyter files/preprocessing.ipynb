```python
import glob
import numpy as np
import pandas as pd
```

```python
root = 'C:\\Users\\Krishna\\Jupyter notebook files\\Molecular Data Science and Informatics\\suppl\\data'
files = sorted(glob.glob('%s\\*.poscar'%root))
```

```python
files
```

```python
predictors = []
at_en = []
egap = []
eps_elec = []
eps_ion = []
eps_tot = []

for i in range(215):
    
    with open(files[i]) as f:
        string = f.read()
    list_str = string.split('\n')
    
    target_list = list_str[0].split(' ')
    dummy=[]
    for j in target_list:
        if j != '':
            dummy.append(j)
    print(dummy[1:6])
    at_en.append(float(dummy[1]))
    egap.append(float(dummy[2]))
    eps_elec.append(float(dummy[3]))
    eps_ion.append(float(dummy[4]))
    eps_tot.append(float(dummy[5]))
    
    pred_list = list_str[6].split(' ')
    dummy=[]
    for j in pred_list:
        if j != '':
            dummy.append(j)
    pred_list = [float(j) for j in dummy]
    predictors.append(pred_list)
```
