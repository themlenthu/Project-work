## Fingerprinting

```python
import os
import pandas as pd
```

```python
root = os.getcwd()
data = root + '\\DataSet2.xlsx'
```

```python
root
```

```python
data_xlsx = pd.read_excel(data)
```

```python
data_xlsx
```

```python
import rdkit
from rdkit.Chem.Descriptors import CalcMolDescriptors
from rdkit.Chem import MolFromSmiles

mol_desc = []

for smiles in data_xlsx['SMILES String']:
    mol = MolFromSmiles(smiles)
    desc = CalcMolDescriptors(mol)
    mol_desc.append(desc)
```

```python
df_desc = pd.DataFrame(mol_desc)
```

```python
df_desc
```

```python
# drop non physico-chemical descriptors

cols = df_desc.columns
cols_nonphy = []
for des in cols:
    if des.startswith('fr_'):
        cols_nonphy.append(des)
df_desc['Glass Transition Temperature'] = data_xlsx['Glass Transition Temperature']
df_desc = df_desc.drop(cols_nonphy, axis=1)
df_desc = df_desc.dropna(axis = 0)
```

```python
df_desc
```

## Model building

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import Model
from tensorflow.keras import Sequential
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
```

```python
Data = np.array(df_desc)
y = np.array(df_desc['Glass Transition Temperature']).reshape((791,1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(Data, y, test_size=0.2)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
```

```python
n_cols = X_train.shape[1]

model = Sequential()
enc_1 = Dense(50, activation='relu', input_shape = (n_cols,))
model.add(enc_1)
enc_2 = Dense(20, activation='relu')
model.add(enc_2)
enc_3 = Dense(10, activation='relu')
model.add(enc_3)
enc_4 = Dense(1, activation='relu')
model.add(enc_4)

lr = 1e-2
optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
model.compile(optimizer=optimizer, loss='mean_squared_error')

dnn = model.fit(X_train, y_train, epochs=50, verbose=1, validation_split=0.01)
```

```python
plt.plot(dnn.history['loss'], label='loss')
plt.plot(dnn.history['val_loss'], label='val_loss')
plt.title("Losses vs Epoch")
plt.xlabel("Epoch")
plt.ylabel("Losses")
plt.grid()
plt.legend()
plt.show()
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_pred_train, y_train))
print(r2_score(y_pred_test,y_test))
```

```python
plt.scatter(y_train, y_pred_train,label='Training Data')
plt.title("Actual Train Data vs Predicted Train Data")
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
plt.legend()
plt.plot(y_train, y_train, 'k-')
plt.grid()
plt.show()

plt.scatter(y_test, y_pred_test, label='Testing Data')
plt.title("Actual Test Data vs Predicted Test Data")
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
plt.legend()
plt.plot(y_train, y_train, 'k-')
plt.grid()
plt.show()

plt.scatter(y_train, y_pred_train,label='Training Data')
plt.scatter(y_test, y_pred_test, label='Testing Data')
plt.title("Actual Data vs Predicted Data")
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
plt.legend()
plt.plot(y_train, y_train, 'k-')
plt.grid()
plt.show()
```
